### Sobre o Conjunto de Dados

**Contexto**  
A análise de recursos humanos (HR Analytics) está revolucionando a forma como os departamentos de recursos humanos operam, proporcionando maior eficiência e melhores resultados. Embora a área de RH utilize análises há anos, a coleta, o processamento e a análise dos dados têm sido, em grande parte, feitos manualmente. Dada a dinâmica dos recursos humanos e os KPIs (indicadores-chave de desempenho) específicos de RH, essa abordagem tradicional limitou o potencial do setor. Por isso, é surpreendente que os departamentos de RH só tenham despertado para a utilidade do aprendizado de máquina mais recentemente. Este conjunto de dados oferece uma oportunidade de experimentar a análise preditiva para identificar os colaboradores mais propensos a serem promovidos.

**Conteúdo**  
O cliente, uma grande multinacional, possui 9 grandes verticais em sua organização. Um dos problemas enfrentados é identificar as pessoas certas para promoção (somente para posições de gerente e abaixo) e prepará-las a tempo. Atualmente, o processo seguido é o seguinte:

1. Primeiramente, é identificado um grupo de funcionários com base em recomendações e no desempenho passado.
2. Os funcionários selecionados passam por um programa de treinamento e avaliação específico para cada vertical. Esses programas são definidos com base nas habilidades exigidas em cada área.
3. Ao final do programa, considerando diversos fatores como o desempenho no treinamento e a conclusão dos KPIs (apenas funcionários com mais de 60% dos KPIs concluídos são elegíveis), os funcionários podem ser promovidos.

Nesse processo, as promoções finais são anunciadas apenas após a avaliação, o que gera atrasos na transição para os novos cargos. A empresa precisa, portanto, de ajuda para identificar candidatos elegíveis em um ponto específico de verificação, para acelerar o ciclo de promoção.

Foram disponibilizados vários atributos sobre o desempenho passado e atual dos funcionários, bem como dados demográficos. A tarefa é prever se um potencial promovido no ponto de verificação no conjunto de teste será promovido ou não após o processo de avaliação.
